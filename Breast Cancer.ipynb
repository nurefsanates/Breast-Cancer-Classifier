```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier, NeighborhoodComponentsAnalysis, LocalOutlierFactor
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("data.csv")
data.head()
```

```python
#drop columns
data.drop(['Unnamed: 32','id'], inplace=True, axis=1)
```

```python
data= data.rename(columns={"diagnosis": "hedef"})
```

```python
plt.figure()
sns.countplot(data["hedef"])
plt.ylabel("Adet")
plt.title("Veri Seti iyi huylu (B) - Kötü huylu (M) Kanser Hücresi Dağılımı")
plt.show()
print(data.hedef.value_counts())
```

```python
#M=1, B=0
data["hedef"]= [1 if i.strip()=="M" else 0 for i in data.hedef]
data.head(3)
```

```python
print("Verinin uzunluğu: ",len(data))
print("Verinin şekli: ", data.shape)
data.info()
data.describe()
```

```python
#Değerler arasında ilişki
corr_matrix= data.corr()
plt.figure(figsize=(12,8))
sns.clustermap(corr_matrix, annot=True, fmt=".1f", linewidths=.01)
plt.title("Özellikler arasındaki ilişki")
plt.show()
```

```python
#hedef ile iyi huylu kötü huylu korelasyonu
threshold=0.75
filter=np.abs(corr_matrix["hedef"])> threshold
corr_features=corr_matrix.columns[filter].tolist()
plt.figure()
sns.clustermap(data[corr_features].corr(), annot=True, fmt=".2f")
plt.title("Özellikler arasındaki ilişki - 0.75 eşik değeri ile")
plt.show()
```

```python
#hedef ile ilişkili özelliklerin saçılımına bakalım
plt.figure()
sns.pairplot(data[corr_features], diag_kind="kde", markers="+", hue="hedef")
plt.show()
```

```python
y=data.hedef
x=data.drop(["hedef"], axis=1)
columns=x.columns.tolist()
```

```python
#yerel aykırı faktör yöntemi
clf= LocalOutlierFactor()
y_pred=clf.fit_predict(x)
#skor
X_score=clf.negative_outlier_factor_
```

```python
#skoru dataframe'e atalım
outlier_score=pd.DataFrame()
outlier_score["score"]=X_score
```

```python
#belirli bir eşik değerine göre aykırı değerler
threshold= -2.2
filtre=outlier_score["score"]<threshold
outlier_index= outlier_score[filtre].index.tolist()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x.iloc[outlier_index,0], x.iloc[outlier_index,1], color="blue", s=50, label="Aykırılar")
plt.scatter(x.iloc[:,0],x.iloc[:,1],color="k",s=3, label="Veri Noktaları")
radius=(X_score.max()-X_score)/(X_score.max()-X_score.min())
outlier_score["radius"]=radius
plt.scatter(x.iloc[:,0], x.iloc[:,1], s=1000*radius, edgecolors="r",alpha=0.5,facecolors="none", label="Aykırı Değer Skoru")
plt.legend()
plt.show()
```

```python
#aykırı değerleri çıkartıyoruz
x=x.drop(outlier_index)
y=y.drop(outlier_index).values
```

```python
#Eğitim
X_egitim, X_test, Y_egitim, Y_test= train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
#standardizasyon
scaler=StandardScaler()
X_egitim=scaler.fit_transform(X_egitim) #eğitim veri setine göre
X_test=scaler.transform(X_test) 
```

```python
X_egitim_df=pd.DataFrame(X_egitim,columns=columns)
X_egitim_df_describe= X_egitim_df.describe()
X_egitim_df["hedef"]=Y_egitim
```

```python
#KUTU GRAFİĞİ
data_melted= pd.melt(X_egitim_df, id_vars="hedef", var_name="features", value_name="value") #uzun formata dönüştürür
plt.figure(figsize=(12,8))
sns.boxplot(x="features", y="value", hue="hedef",data=data_melted)
plt.xticks(rotation=90)
plt.show()
```

```python
#model
knn= KNeighborsClassifier(n_neighbors=2)
#eğitim
knn.fit(X_egitim,Y_egitim)
```

```python
#tahmin
y_pred=knn.predict(X_test)
```

```python
#karışıklık matrisi
cm=confusion_matrix(Y_test,y_pred)
print("Karışıklık Matrisi: ",cm)
plt.figure()
sns.heatmap(cm,annot=True,fmt=".0f")
plt.show()
```

```python
#accuracy
dogruluk=accuracy_score(Y_test,y_pred)
print("Doğruluk: ",dogruluk)
```

```python
#en iyi k değeri
def eniyiParametreler(x_egitim, x_test, y_egitim, y_test):
    k_range=list(range(1,31))
    #uniform: tüm noktalar eşit ağırlık
    #distance: mesafelerine göre ağırlık
    weight_options=["uniform","distance"]
    param_grid=dict(n_neighbors=k_range, weights=weight_options)
    #knn sınıflandırıcı
    knn=KNeighborsClassifier()
    #çapraz doğrulama ile
    grid=GridSearchCV(knn, param_grid, cv=10, scoring="accuracy")
    #eğitim
    grid.fit(x_egitim,y_egitim)

    print(f"En iyi eğitim skoru: {grid.best_score_} ve parametreler {grid.best_params_}")
    print()
    #en iyi sonucu veren parametreler ile yeni bir model
    knn=KNeighborsClassifier(**grid.best_params_)
    #eğitim
    knn.fit(x_egitim, y_egitim)

    #tahminleme
    y_pred_test= knn.predict(x_test)
    y_pred_train=knn.predict(x_egitim)
    #karışıklık matrisi
    cm_test=confusion_matrix(y_test,y_pred_test)
    cm_train=confusion_matrix(y_egitim,y_pred_train)
    #doğruluk ve sonuç yazdırılması
    acc_test=accuracy_score(y_test,y_pred_test)
    acc_train=accuracy_score(y_egitim,y_pred_train)
    print(f"Test doğruluğu: {acc_test}, Eğitim doğruluğu: {acc_train}")
    print()
    print("Karışıklık matrisi Test: ",cm_test)
    print("Karışıklık matrisi eğitim: ",cm_train)
    return grid
```

```python
grid=eniyiParametreler(X_egitim, X_test, Y_egitim, Y_test)
```

```python
#VERİMİZİ 2 TEMEL BİLEŞENE DÜŞÜRME
#standardizasyon
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
#temel bileşeni tanımla, 2 tane bileşen
pca=PCA(n_components=2)
```

```python
#pca eğitimi
pca.fit(x_scaled)
```

```python
#pca uygula
X_reduced_pca=pca.transform(x_scaled)
```

```python
pca_data=pd.DataFrame(X_reduced_pca,columns=["p1","p2"])
pca_data["hedef"]=y
plt.figure()
sns.scatterplot(x="p1",y="p2", hue="hedef", data=pca_data)
plt.title("Temel Veri Analizi")
plt.xlabel("Bileşen 1")
plt.ylabel("Bileşen 2")
plt.show()
```

```python
#pca sonucu elde edilen verinin eğitim ve test bölünmesi
X_train_pca,X_test_pca,Y_train_pca,Y_test_pca= train_test_split(X_reduced_pca, y, test_size=0.3, random_state=42)
```

```python
#pca sonucu elde edilen verilen en iyi knn değerlerinin bulunması
grid_pca= eniyiParametreler(X_train_pca,X_test_pca,Y_train_pca,Y_test_pca)
```

```python
#pca sonucu elde edilen verinin tahminlenmesi ve görselleştirilmesi
cmap_light=ListedColormap(['orange','cornflowerblue'])
cmap_bold=ListedColormap(['darkorange','darkblue'])

h=0.5 #mesh boyutu
X=X_reduced_pca
#görselleştirmenin sınırlarını genişletme
x_min, x_max= X[:,0].min()-1, X[:,0].max()+1
y_min, y_max= X[:,1].min()-1, X[:,1].max()+1
xx,yy=np.meshgrid(np.arange(x_min, x_max,h), np.arange(y_min,y_max,h))

Z=grid_pca.predict(np.c_[xx.ravel(), yy.ravel()])

Z=Z.reshape(xx.shape)
plt.figure(figsize=(12,8))
plt.pcolormesh(xx,yy,Z,cmap=cmap_light)

plt.scatter(X[:,0],X[:,1],c=y, cmap=cmap_bold,edgecolor='k',s=20)
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title(f"{2} - Etiket Sınıflandırması: k={grid_pca.best_estimator_.n_neighbors}, ağırlıklar={grid_pca.best_estimator_.weights}")
plt.show()
```

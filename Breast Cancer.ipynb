```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier, NeighborhoodComponentsAnalysis, LocalOutlierFactor
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("data.csv")
data.head()
```

```python
#drop columns
data.drop(['Unnamed: 32','id'], inplace=True, axis=1)
```

```python
data= data.rename(columns={"diagnosis": "hedef"})
```

```python
plt.figure()
sns.countplot(data["hedef"])
plt.ylabel("Adet")
plt.title("Veri Seti iyi huylu (B) - KÃ¶tÃ¼ huylu (M) Kanser HÃ¼cresi DaÄŸÄ±lÄ±mÄ±")
plt.show()
print(data.hedef.value_counts())
```

```python
#M=1, B=0
data["hedef"]= [1 if i.strip()=="M" else 0 for i in data.hedef]
data.head(3)
```

```python
print("Verinin uzunluÄŸu: ",len(data))
print("Verinin ÅŸekli: ", data.shape)
data.info()
data.describe()
```

```python
#DeÄŸerler arasÄ±nda iliÅŸki
corr_matrix= data.corr()
plt.figure(figsize=(12,8))
sns.clustermap(corr_matrix, annot=True, fmt=".1f", linewidths=.01)
plt.title("Ã–zellikler arasÄ±ndaki iliÅŸki")
plt.show()
```

```python
#hedef ile iyi huylu kÃ¶tÃ¼ huylu korelasyonu
threshold=0.75
filter=np.abs(corr_matrix["hedef"])> threshold
corr_features=corr_matrix.columns[filter].tolist()
plt.figure()
sns.clustermap(data[corr_features].corr(), annot=True, fmt=".2f")
plt.title("Ã–zellikler arasÄ±ndaki iliÅŸki - 0.75 eÅŸik deÄŸeri ile")
plt.show()
```

```python
#hedef ile iliÅŸkili Ã¶zelliklerin saÃ§Ä±lÄ±mÄ±na bakalÄ±m
plt.figure()
sns.pairplot(data[corr_features], diag_kind="kde", markers="+", hue="hedef")
plt.show()
```

```python
y=data.hedef
x=data.drop(["hedef"], axis=1)
columns=x.columns.tolist()
```

```python
#yerel aykÄ±rÄ± faktÃ¶r yÃ¶ntemi
clf= LocalOutlierFactor()
y_pred=clf.fit_predict(x)
#skor
X_score=clf.negative_outlier_factor_
```

```python
#skoru dataframe'e atalÄ±m
outlier_score=pd.DataFrame()
outlier_score["score"]=X_score
```

```python
#belirli bir eÅŸik deÄŸerine gÃ¶re aykÄ±rÄ± deÄŸerler
threshold= -2.2
filtre=outlier_score["score"]<threshold
outlier_index= outlier_score[filtre].index.tolist()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x.iloc[outlier_index,0], x.iloc[outlier_index,1], color="blue", s=50, label="AykÄ±rÄ±lar")
plt.scatter(x.iloc[:,0],x.iloc[:,1],color="k",s=3, label="Veri NoktalarÄ±")
radius=(X_score.max()-X_score)/(X_score.max()-X_score.min())
outlier_score["radius"]=radius
plt.scatter(x.iloc[:,0], x.iloc[:,1], s=1000*radius, edgecolors="r",alpha=0.5,facecolors="none", label="AykÄ±rÄ± DeÄŸer Skoru")
plt.legend()
plt.show()
```

```python
#aykÄ±rÄ± deÄŸerleri Ã§Ä±kartÄ±yoruz
x=x.drop(outlier_index)
y=y.drop(outlier_index).values
```

```python
#EÄŸitim
X_egitim, X_test, Y_egitim, Y_test= train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
#standardizasyon
scaler=StandardScaler()
X_egitim=scaler.fit_transform(X_egitim) #eÄŸitim veri setine gÃ¶re
X_test=scaler.transform(X_test) 
```

```python
X_egitim_df=pd.DataFrame(X_egitim,columns=columns)
X_egitim_df_describe= X_egitim_df.describe()
X_egitim_df["hedef"]=Y_egitim
```

```python
#KUTU GRAFÄ°ÄžÄ°
data_melted= pd.melt(X_egitim_df, id_vars="hedef", var_name="features", value_name="value") #uzun formata dÃ¶nÃ¼ÅŸtÃ¼rÃ¼r
plt.figure(figsize=(12,8))
sns.boxplot(x="features", y="value", hue="hedef",data=data_melted)
plt.xticks(rotation=90)
plt.show()
```

```python
#model
knn= KNeighborsClassifier(n_neighbors=2)
#eÄŸitim
knn.fit(X_egitim,Y_egitim)
```

```python
#tahmin
y_pred=knn.predict(X_test)
```

```python
#karÄ±ÅŸÄ±klÄ±k matrisi
cm=confusion_matrix(Y_test,y_pred)
print("KarÄ±ÅŸÄ±klÄ±k Matrisi: ",cm)
plt.figure()
sns.heatmap(cm,annot=True,fmt=".0f")
plt.show()
```

```python
#accuracy
dogruluk=accuracy_score(Y_test,y_pred)
print("DoÄŸruluk: ",dogruluk)
```

```python
#en iyi k deÄŸeri
def eniyiParametreler(x_egitim, x_test, y_egitim, y_test):
    k_range=list(range(1,31))
    #uniform: tÃ¼m noktalar eÅŸit aÄŸÄ±rlÄ±k
    #distance: mesafelerine gÃ¶re aÄŸÄ±rlÄ±k
    weight_options=["uniform","distance"]
    param_grid=dict(n_neighbors=k_range, weights=weight_options)
    #knn sÄ±nÄ±flandÄ±rÄ±cÄ±
    knn=KNeighborsClassifier()
    #Ã§apraz doÄŸrulama ile
    grid=GridSearchCV(knn, param_grid, cv=10, scoring="accuracy")
    #eÄŸitim
    grid.fit(x_egitim,y_egitim)

    print(f"En iyi eÄŸitim skoru: {grid.best_score_} ve parametreler {grid.best_params_}")
    print()
    #en iyi sonucu veren parametreler ile yeni bir model
    knn=KNeighborsClassifier(**grid.best_params_)
    #eÄŸitim
    knn.fit(x_egitim, y_egitim)

    #tahminleme
    y_pred_test= knn.predict(x_test)
    y_pred_train=knn.predict(x_egitim)
    #karÄ±ÅŸÄ±klÄ±k matrisi
    cm_test=confusion_matrix(y_test,y_pred_test)
    cm_train=confusion_matrix(y_egitim,y_pred_train)
    #doÄŸruluk ve sonuÃ§ yazdÄ±rÄ±lmasÄ±
    acc_test=accuracy_score(y_test,y_pred_test)
    acc_train=accuracy_score(y_egitim,y_pred_train)
    print(f"Test doÄŸruluÄŸu: {acc_test}, EÄŸitim doÄŸruluÄŸu: {acc_train}")
    print()
    print("KarÄ±ÅŸÄ±klÄ±k matrisi Test: ",cm_test)
    print("KarÄ±ÅŸÄ±klÄ±k matrisi eÄŸitim: ",cm_train)
    return grid
```

```python
grid=eniyiParametreler(X_egitim, X_test, Y_egitim, Y_test)
```

```python
#VERÄ°MÄ°ZÄ° 2 TEMEL BÄ°LEÅžENE DÃœÅžÃœRME
#standardizasyon
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
#temel bileÅŸeni tanÄ±mla, 2 tane bileÅŸen
pca=PCA(n_components=2)
```

```python
#pca eÄŸitimi
pca.fit(x_scaled)
```

```python
#pca uygula
X_reduced_pca=pca.transform(x_scaled)
```

```python
pca_data=pd.DataFrame(X_reduced_pca,columns=["p1","p2"])
pca_data["hedef"]=y
plt.figure()
sns.scatterplot(x="p1",y="p2", hue="hedef", data=pca_data)
plt.title("Temel Veri Analizi")
plt.xlabel("BileÅŸen 1")
plt.ylabel("BileÅŸen 2")
plt.show()
```

```python
#pca sonucu elde edilen verinin eÄŸitim ve test bÃ¶lÃ¼nmesi
X_train_pca,X_test_pca,Y_train_pca,Y_test_pca= train_test_split(X_reduced_pca, y, test_size=0.3, random_state=42)
```

```python
#pca sonucu elde edilen verilen en iyi knn deÄŸerlerinin bulunmasÄ±
grid_pca= eniyiParametreler(X_train_pca,X_test_pca,Y_train_pca,Y_test_pca)
```

```python
#pca sonucu elde edilen verinin tahminlenmesi ve gÃ¶rselleÅŸtirilmesi
cmap_light=ListedColormap(['orange','cornflowerblue'])
cmap_bold=ListedColormap(['darkorange','darkblue'])

h=0.5 #mesh boyutu
X=X_reduced_pca
#gÃ¶rselleÅŸtirmenin sÄ±nÄ±rlarÄ±nÄ± geniÅŸletme
x_min, x_max= X[:,0].min()-1, X[:,0].max()+1
y_min, y_max= X[:,1].min()-1, X[:,1].max()+1
xx,yy=np.meshgrid(np.arange(x_min, x_max,h), np.arange(y_min,y_max,h))

Z=grid_pca.predict(np.c_[xx.ravel(), yy.ravel()])

Z=Z.reshape(xx.shape)
plt.figure(figsize=(12,8))
plt.pcolormesh(xx,yy,Z,cmap=cmap_light)

plt.scatter(X[:,0],X[:,1],c=y, cmap=cmap_bold,edgecolor='k',s=20)
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title(f"{2} - Etiket SÄ±nÄ±flandÄ±rmasÄ±: k={grid_pca.best_estimator_.n_neighbors}, aÄŸÄ±rlÄ±klar={grid_pca.best_estimator_.weights}")
plt.show()
```
